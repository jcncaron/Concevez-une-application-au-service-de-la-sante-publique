## Idée d'application :

* PureScore :  
  -> Manger mieux (nutriments et énergie),  
  -> Manger des aliments non ou moins transformés (nombre d'additifs - pas pris en compte dans nutriscore!),  
  -> Manger des aliments sans additifs ou avec moins d'additifs (nova group - pas pris en compte dans nutriscore!)  
  
  
* Variables nécessaires à l'application :  
  -> Code : pour repérer et supprimer les duplicatas  
  -> url : pour vérifier certains produits au cas par cas 
  -> image_url : pour le rendu visuel de l'appli  
  -> product_name : pour le rendu visuel de l'appli  
  -> cetegories_en, main_category_en, pnns_groups_1 & pnns_groups_2 : pour classer les produits et proposer des alternatives par catégorie  
  -> countries_en : pour filtrer les produits par pays  
  -> additives_n : pour calculer un score 'additifs'  
  -> nutriscore_score, nutriscore_grade : pour évaluer les corrélations avec le PureScore et les autres variables  
  -> nova_group : classement selon le degré de transformation du produit - pour calculer un score 'nova'  
  -> energy, fat, saturated_fat, carbohydrates, sugars, proteins, salt, fiber, fruits_vegetables_nuts : pour calculer un score 'nutritive   
   
   
* Scoring envisagé :  note / 10 au total  
  -> 3 points pour score 'additifs  
  -> 3 points pour score 'nova'  
  -> 4 points pour score 'nutriments'    
  
  
* Produits alternatifs proposés suite au scan d'un produit :   
  -> produits appartenant à des groupes plus ou moins restreints avec un PureScore supérieur à celui du produit scanné
  

### Intérêt : 
valoriser les aliments sans additifs et non transformés (plus sain) tout en prenant en compte les nutriments(qui rentrent en compte dans le calcul du nutriscore)

```python
# Création d'une fonction pour supprimer les duplicatas
def remove_duplicates(df):
    # on compte le nombre de valeurs manquantes par ligne, pour savoir quel duplicata supprimer
    # et on stocke les données supprimées dans une nouvelle colonne
    df['NB_NAN']=df.isna().sum(axis=1)
    # on trie par ordre croissant les lignes en fonction du nombre de valeurs manquantes (à l'aide de la nouvelle colonne créée)
    df = df.sort_values('NB_NAN')
    # suppression des duplicatas en gardant les versions les plus remplies
    df = df.drop_duplicates('code', keep='first')
    # on supprime la colonne qui n'est plus utile
    df = df.drop('NB_NAN', axis=1)
    return df
```

```python
# Comme ci-dessus, on attribue les bonnes catégories "Unsweetened beverages"
# ou "Artificially sweetened beverages" à la place de "Alcoholic beverages"
# Pour les pnns_groups_2
def alcoholic_to_unalcoholic(df):
    df['pnns_groups_2'].loc[(df['pnns_groups_2']=='Alcoholic beverages')
                            & (df['sugars_100g'] > 10)] = 'Artificially sweetened beverages'
    df['pnns_groups_2'].loc[(df['pnns_groups_2']=='Alcoholic beverages')
                            & (df['sugars_100g'] < 10)] = 'Unsweetened beverages'
    return df
```

```python
# Fonction pour remplacer les 'unknown' par des pnns_groups_2
def replace_unknown_by_pnns_2(df):
    # On enregistre l'indice de position de la colonne "pnns_groups_2"
    index_pnns_2 = df.columns.get_loc("pnns_groups_2")
    # On utilise l'index de position de la list_pnns_groups_2
    for i in range(len(list_pnns_groups_2)):
        # On enregistre la liste des indices de ligne pour lesquels on a une correspondance main_category/pnns_groups_2
        index_list = df.index[(df['main_category_en'].isin((list(dict_main_category_by_pnns_2.values())[i])))].tolist()
        for row_index in index_list:
            # Si 'unknown' est contenu dans la colonne, on le remplace par le pnns_groups_2 correspondant à la main_category
            if df.iloc[row_index,index_pnns_2] == 'unknown':
                df.iloc[row_index,index_pnns_2] = df.iloc[row_index,index_pnns_2].replace('unknown', list_pnns_groups_2[i])
            # Si un pnns_groups_2 est déjà attribué, on ne change rien
            else:
                continue
    return df
```

```python
# Fonction pour remplacer les 'unknown' par des pnns_groups_2
# en inversant l'ordre des boucles
def replace_unknown_by_pnns_2_reverse(df):
    for index in df.index:
        for j in range(len(list_pnns_groups_2)):
            df['pnns_groups_2'].loc[df['main_category_en'].isin((list(dict_main_category_by_pnns_2.values())[j]))] = list_pnns_groups_2[j]
    return df
# traitement beaucoup plus long que la fonction "replace_unknown_by_pnns_2" !
```

```python
# Nouvelle fonction simplifiée sans faire appel aux indices de lignes
def replace_unknown_by_pnns_group_2(df):
    for i in range(len(list_pnns_groups_2)):
        df['pnns_groups_2'].loc[df['main_category_en'].isin((list(dict_main_category_by_pnns_2.values())[i]))] = list_pnns_groups_2[i]
    return df
# apply sur cette fonction : traitement très long
```

```python
# Fonction permettant de remplacer les valeurs NaN
# des différentes colonnes nutrition ou energy
# par la médiane de chaque colonne par pnns_group_2

list_col_median = ['energy_100g',
                   'fat_100g',
                   'saturated_fat_100g',
                   'carbohydrates_100g',
                   'sugars_100g',
                   'proteins_100g',
                   'salt_100g',
                   'fiber_100g',
                   'fruits_vegetables_nuts_estimate_from_ingredients_100g']

def replace_NaN_by_median_pnns_2(df):
    for col in list_col_median:
        # indexation des colonnes de la liste définie ci-dessus
        index_col = df.columns.get_loc(col)
        # indexation des lignes contenant des NaN pour les colonnes de la liste ci-dessus
        index_row_nan = df.loc[df[col].isna()].index
        for i in range(len(list_pnns_groups_2)):
            # Calcul de médiane pour chaque colonne et chaque catégorie pnns_groups_2
            # (en ne comptabilisant pas les NaN)
            med = df[col].loc[df['pnns_groups_2'] == list_pnns_groups_2[i]].median(skipna = True)
            # Attribution de la valeur de la médiane à la place des NaN correspondant
            df.iloc[index_row_nan,index_col] = med
    return df
```

```python
def nettoyage(df, country):
    
    # Création de la liste des colonnes conservées pour l'application
    list_columns_for_application = ['code',
                                    'url',
                                    'image_url',
                                    'product_name',
                                    'categories_en',
                                    'countries_en',
                                    'additives_n',
                                    'nutriscore_score',
                                    'nutriscore_grade',
                                    'nova_group',
                                    'pnns_groups_1',
                                    'pnns_groups_2',
                                    'main_category_en',
                                    'energy_100g',
                                    'fat_100g',
                                    'saturated-fat_100g',
                                    'carbohydrates_100g',
                                    'sugars_100g',
                                    'proteins_100g',
                                    'salt_100g',
                                    'fiber_100g',
                                    'fruits-vegetables-nuts-estimate-from-ingredients_100g']
    
    # Création d'une fonction permettant de sélectionner
    # les colonnes utiles pour l'application "manger mieux, manger sain"
    # via une liste de colonnes
    def filter_columns(df):
        df = df[list_columns_for_application]
        return df
    
    df = filter_columns(df)
    
    # On remplace les tirets "-" par des underscores "_" dans les noms des variables concernées
    df.rename(columns={'saturated-fat_100g': 'saturated_fat_100g',
                       'fruits-vegetables-nuts-estimate-from-ingredients_100g':
                       'fruits_vegetables_nuts_estimate_from_ingredients_100g'},
              inplace=True)
    
    # Création d'une fonction pour filtrer le dataframe par pays
    def country_filter(df, country):
        df = df.loc[df["countries_en"].str.contains(country, na=False)]
        return df
    
    df = country_filter(df, "France")
    
    # Création d'une liste de colonne pour lesquelles les NaN
    # ne peuvent pas être remplacés
    list_not_alterable_columns = ['image_url',
                              'nova_group',
                              'additives_n',
                              'product_name',
                              'nutriscore_grade',
                              'nutriscore_score',
                              'categories_en',
                              'main_category_en']
    
    # Création d'une fonction pour supprimer les lignes contenant des NaN
    # en sélectionnant certaines colonnes uniquement
    def delete_NaN(df):
        df = df.dropna(subset = list_not_alterable_columns)
        return df
   
    df = delete_NaN(df)
    
    # Création d'une liste contenant toutes les colonnes
    # qui doivent avoir des valeurs comprises entre 0 et 100
    list_nutrition_columns = ['fat_100g',
                              'saturated_fat_100g',
                              'carbohydrates_100g',
                              'sugars_100g',
                              'proteins_100g',
                              'salt_100g',
                              'fiber_100g',
                              'fruits_vegetables_nuts_estimate_from_ingredients_100g']
    
    # Création d'une fonction pour supprimer les lignes contenant
    # des valeurs aberrantes (> 100) pour une sélection de colonnes
    def remove_outlier(df, col_list):
        for col in col_list:
            filtre = df[(df[col] > 100)
                        | (df[col] < 0)].index
            df = df.drop(filtre)
        return(df)
    
    df = remove_outlier(df, list_nutrition_columns)
    
    # Création de listes de colonnes à sommer
    # afin de supprimer des lignes contenant des valeurs aberrantes
    list_columns_to_sum_1 = ['saturated_fat_100g',
                             'sugars_100g',
                             'proteins_100g',
                             'salt_100g']

    list_columns_to_sum_2 = ['fat_100g',
                             'sugars_100g',
                             'proteins_100g',
                             'salt_100g']

    list_columns_to_sum_3 = ['fat_100g',
                             'carbohydrates_100g',
                             'proteins_100g',
                             'salt_100g']

    list_columns_to_sum_4 = ['saturated_fat_100g',
                             'carbohydrates_100g',
                             'proteins_100g',
                             'salt_100g']
    
    # Création d'une fonction pour suppression des lignes aberrantes
    # (somme des nutriments > 100g pour une portion de 100g)
    def remove_outliers_rows(df):
        df = df.loc[(df[list_columns_to_sum_1].sum(axis=1)) <=100]
        df = df.loc[(df[list_columns_to_sum_2].sum(axis=1)) <=100]
        df = df.loc[(df[list_columns_to_sum_3].sum(axis=1)) <=100]
        df = df.loc[(df[list_columns_to_sum_4].sum(axis=1)) <=100]
        return df
    
    df = remove_outliers_rows(df)
    
    # Création d'une fonction pour supprimer les lignes contenant
    # des valeurs aberrantes (> 3807) pour la variable "energy_100g"
    def remove_outlier_energy(df):
        filtre = df[df['energy_100g'] > 3807].index
        df = df.drop(filtre)
        return(df)

    df = remove_outlier_energy(df)
    
    # Création d'une fonction pour supprimer les duplicatas
    def remove_duplicates(df):
        # on compte le nombre de valeurs manquantes par ligne, pour savoir quel duplicata supprimer
        # et on stocke les données supprimées dans une nouvelle colonne
        df['NB_NAN']=df.isna().sum(axis=1)
        # on trie par ordre croissant les lignes en fonction du nombre de valeurs manquantes (à l'aide de la nouvelle colonne créée)
        df = df.sort_values('NB_NAN')
        # suppression des duplicatas en gardant les versions les plus remplies
        df = df.drop_duplicates('code', keep='first')
        # on supprime la colonne qui n'est plus utile
        df = df.drop('NB_NAN', axis=1)
        return df
    
    df = remove_duplicates(df)
    
    # On remplace 'Alcoholic beverages' par 'Beverages'
    mydata_france['pnns_groups_1'].loc[mydata_france['pnns_groups_1']=='Alcoholic beverages'] = 'Beverages'
    
    # On attribue les bonnes catégories "Unsweetened beverages"
    # ou "Artificially sweetened beverages" à la place de "Alcoholic beverages"
    # Pour les pnns_groups_2
    def alcoholic_to_unalcoholic(df):
        df['pnns_groups_2'].loc[(df['pnns_groups_2']=='Alcoholic beverages')
                                & (df['sugars_100g'] > 10)] = 'Artificially sweetened beverages'
        df['pnns_groups_2'].loc[(df['pnns_groups_2']=='Alcoholic beverages')
                                & (df['sugars_100g'] < 10)] = 'Unsweetened beverages'
        return df
    
    df = alcoholic_to_unalcoholic(df)
    
    # Création d'une liste contenant les catégories pnns_groups_1 
    # (sans la catégorie 'unknown')
    list_pnns_groups_1 = df["pnns_groups_1"].unique().tolist()
    list_pnns_groups_1.remove('unknown')
    list_pnns_groups_1

    # Création d'une liste contenant les catégories pnns_groups_2
    # (sans la catégorie 'unknown')
    list_pnns_groups_2 = df["pnns_groups_2"].unique().tolist()
    list_pnns_groups_2.remove('unknown')

    # Création d'un dictionnaire qui regroupe les groupes pnns 2 par groupe pnns 1
    dict_pnns_2_by_pnns_1 = {"list_" + pnns_1.replace(" ", "_").replace("-", "_").lower() + "_pnns_groups_2":
                         (df.loc[df["pnns_groups_1"] == pnns_1]["pnns_groups_2"].unique().tolist())
                         for pnns_1 in list_pnns_groups_1}

    # Création d'un dictionnaire qui regroupe les categories_en par groupes pnns 2
    dict_category_by_pnns_2 = {"list_" + pnns_2.replace(" ", "_").replace("-", "_").lower() + "_category":
                         (df.loc[df["pnns_groups_2"] == pnns_2]["categories_en"].unique().tolist())
                         for pnns_2 in list_pnns_groups_2}

    # Création d'un dictionnaire qui regroupe les main_categories_en dans les groupes pnns 2
    dict_main_category_by_pnns_2 = {"list_" + pnns_2.replace(" ", "_").replace("-", "_").lower() + "_main_category":
                                (df.loc[df["pnns_groups_2"] == pnns_2]["main_category_en"].unique().tolist())
                                for pnns_2 in list_pnns_groups_2}
    
    mydata_unknown = df.loc[df['pnns_groups_2'] == 'unknown']
    mydata_unknown = mydata_unknown.reset_index(drop=False)
    
    # Fonction pour remplacer les 'unknown' par des pnns_groups_2
    def replace_unknown_by_pnns_2():
        # On enregistre l'indice de position de la colonne "pnns_groups_2"
        index_pnns_2 = mydata_unknown.columns.get_loc("pnns_groups_2")
        # On utilise l'index de position de la list_pnns_groups_2
        for i in range(len(list_pnns_groups_2)):
            # On enregistre la liste des indices de ligne pour lesquels on a une correspondance main_category/pnns_groups_2
            index_list = mydata_unknown.index[(mydata_unknown['main_category_en'].isin((list(dict_main_category_by_pnns_2.values())[i])))].tolist()
            for row_index in index_list:
                # Si 'unknown' est contenu dans la colonne, on le remplace par le pnns_groups_2 correspondant à la main_category
                if mydata_unknown.iloc[row_index,index_pnns_2] == 'unknown':
                    mydata_unknown.iloc[row_index,index_pnns_2] = mydata_unknown.iloc[row_index,index_pnns_2].replace('unknown', list_pnns_groups_2[i])
                # Si un pnns_groups_2 est déjà attribué, on ne change rien
                else:
                    continue
        return mydata_unknown
    
    mydata_unknown = replace_unknown_by_pnns_2()
    
    # Fonction pour remplacer les 'unknown' par des pnns_groups_1
    def replace_unknown_by_pnns_1():
    # On obtient l'indice de position de la colonne "pnns_groups_1"
        index_pnns_1 = mydata_unknown.columns.get_loc("pnns_groups_1")
        for i in range(len(list_pnns_groups_1)):
            # On obtient la liste des indices pour lesquels on a une correspondance pnns_groups_2/pnns_groups_1
            index_list = mydata_unknown.index[(mydata_unknown['pnns_groups_2'].isin((list(dict_pnns_2_by_pnns_1.values())[i])))].tolist()
            for row_index in index_list:
                if mydata_unknown.iloc[row_index,index_pnns_1] == 'unknown':
                    mydata_unknown.iloc[row_index,index_pnns_1] = mydata_unknown.iloc[row_index,index_pnns_1].replace('unknown', list_pnns_groups_1[i])
                else:
                    continue
        return mydata_unknown
    
    mydata_unknown = replace_unknown_by_pnns_1()

    # Get names of indexes for which column Stock has value 'unknown' in mydata
    indexNames = df.loc[mydata['pnns_groups_1'] == 'unknown' ].index
    # Delete these row indexes from dataFrame
    df = df.drop(indexNames)
    
    # Get names of indexes for which column Stock has value 'unknown' in mydata_france_unknown
    indexNames = mydata_unknown.loc[mydata_unknown['pnns_groups_1'] == 'unknown' ].index
    # Delete these row indexes from dataFrame
    mydata_unknown = mydata_unknown.drop(indexNames)
    
    # Refusionner les dataframe mydata (sans les unknown) et mydata_unknown (sans les unknown)
    mydata_final = pd.concat([df,mydata_unknown], ignore_index=True)
    mydata_final = mydata_final.drop(['index'], axis=1)
    
    
    # Fonction permettant de remplacer les valeurs NaN
    # des différentes colonnes nutrition ou energy
    # par la médiane de chaque colonne par pnns_group_2

    list_col_median = ['energy_100g',
                   'fat_100g',
                   'saturated_fat_100g',
                   'carbohydrates_100g',
                   'sugars_100g',
                   'proteins_100g',
                   'salt_100g',
                   'fiber_100g',
                   'fruits_vegetables_nuts_estimate_from_ingredients_100g']
    
    def replace_NaN_by_median_pnns_2():
        for col in list_col_median:
            index_col = mydata_final.columns.get_loc(col)
            index_row_nan = mydata_final.loc[mydata_final[col].isna()].index
            for i in range(len(list_pnns_groups_2)):
                med = mydata_final[col].loc[mydata_final['pnns_groups_2'] == list_pnns_groups_2[i]].median(skipna = True)
                mydata_final.iloc[index_row_nan,index_col] = med
        return mydata_final
    
    replace_NaN_by_median_pnns_2()
    
    # renvoyer le dataframe final nettoyé
    return mydata_final
```
